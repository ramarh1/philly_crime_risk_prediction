**Goal:** Forecast crime risk across Philadelphia neighborhoods using operational and public safety data  

## 1. Problem Definition

### Objective
The goal of this project is to predict the likelihood of crime occurring in a given neighborhood and time window using:

- Historical crime incident data
- 311 service request activity

### Business/Operational Value
Understanding when and where crime risk is elevated can help:
- Allocate resources more effectively
- Identify patterns related to environmental or service issues
- Support proactive decision-making

### Key Question
Can 311 request patterns help explain or predict crime activity?

## 2. Data Sources

### 1. Crime Incidents Dataset
- Source: OpenDataPhilly
- Description: Records of reported crimes including type, time, and location
- Time range: 2006–Present

### 2. 311 Service Requests Dataset
- Source: OpenDataPhilly
- Description: Service requests submitted by residents (e.g., sanitation, graffiti, abandoned vehicles)
- Time range: 2014–Present

### Assumptions
- Location data can be mapped to neighborhoods or districts
- Time fields are consistent and usable for aggregation

### Limitations
- Reporting bias in both datasets
- Missing or incomplete location/time entries
