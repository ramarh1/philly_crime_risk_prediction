## 6. Feature Engineering

### Time-Based Features
- Hour of day
- Day of week
- Weekend vs weekday
- Month / season

### Crime Features
- Rolling average of crimes (past 3–7 days)
- Crime counts by type

### 311 Features
- Count of requests by type
- Total request volume
- Lagged request volume (previous time windows)

### Spatial Features
- Neighborhood / district encoding

### Target Variable
- Binary: Crime occurred (yes/no)
OR
- Count: Number of crimes in a time window
