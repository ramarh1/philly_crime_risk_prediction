## 7. Modeling Approach

### Model Goal
Predict crime risk based on historical and operational signals

### Models Considered
- Logistic Regression (baseline)
- Random Forest
- Gradient Boosting (XGBoost)

### Approach
- Train-test split based on time (not random)
- Avoid data leakage by using past data only

### Evaluation Metrics
- Classification:
  - Accuracy
  - Precision / Recall
  - ROC-AUC

- Regression:
  - RMSE
  - MAE
