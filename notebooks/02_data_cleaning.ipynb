## 4. Data Cleaning

### Crime Dataset
- Remove null or invalid timestamps
- Standardize location fields
- Filter to relevant time period

### 311 Dataset
- Drop irrelevant columns
- Normalize request types
- Handle missing geographic data

### Common Cleaning Steps
- Convert date/time fields
- Ensure consistent formatting for joins


## 5. Data Integration

### Strategy
Both datasets are aggregated to the same level before merging:

- Geography: Neighborhood or police district
- Time: Hour, Day, or Week

### Aggregation Examples
- Number of crimes per area per hour
- Number of 311 requests per area per hour

### Merging Approach
- Join datasets on:
  - Location
  - Time window

### Output
A unified dataset combining crime counts and 311 activity

### Handling Multi-Year 311 Data

The 311 Service Requests dataset is provided as separate files for each year.

To ensure consistency and enable time-based analysis across the full dataset,
all yearly files are concatenated into a single dataset prior to cleaning.

This approach allows:
- Standardized preprocessing across all years
- Easier feature engineering
- Improved model reliability

### Time Range Selection

Although data is available from 2014 onward, this analysis focuses on data from 2018–2024.

This range was selected to:
- Ensure data consistency across datasets
- Reduce noise from earlier reporting inconsistencies
- Focus on recent patterns for more relevant predictions

Additionally, a time-based split is used:
- Training: 2018–2022
- Testing: 2023–2024
